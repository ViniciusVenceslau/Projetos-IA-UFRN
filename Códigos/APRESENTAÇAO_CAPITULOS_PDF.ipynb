### **3.**   Data Feature Selection:



Quanto mais próximo de **1 ou -1** o valor, mais forte é a correlação (positiva ou negativa, respectivamente).

Valores próximos de 0 indicam pouca ou nenhuma correlação.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

# Carregar os dados do conjunto de dados wine usando o scikit-learn.
wine_data = load_wine()

# Criar um DataFrame com os dados e os nomes das colunas.
dados = pd.DataFrame(wine_data['data'], columns=wine_data['feature_names'])

# Adicionar a coluna de classe ao DataFrame.
dados['Class'] = wine_data['target']

# Calcular a matriz de correlação.
matriz_correlacao = dados.corr()

# Plotar a matriz de correlação usando um heatmap.
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlação das Características do Wine')
plt.show()
```

### **4.**   Logistic Regression:

Como este método em específico será utlizado para fins de classificação, conforme explicado no material em PDF, será feita a avaliação considerando a temática "**Multinominal**".

Em um tipo de classificação como essa, a variável dependente pode ter 3 ou mais tipos possíveis não ordenados ou tipos que não têm significado quantitativo. Por exemplo, essas variáveis podem representar 'Tipo A', 'Tipo B' ou 'Tipo C'.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Carregar o conjunto de dados de vinhos.
wine = load_wine()
X = wine.data
y = wine.target

# Dividir os dados em conjuntos de treino e teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=13)

# Criar o classificador de Regressão Logística -> "max_iter=10000" representa o número de iterações na tentativa de encontrar uma solução aceitável.
clf = LogisticRegression(max_iter=10000)

# Treinar o modelo.
clf.fit(X_train, y_train)

# Fazer previsões no conjunto de teste.
y_pred = clf.predict(X_test)

# Calcular e exibir a acurácia.
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia da regressão logística: {accuracy:.2f}')
print('\n')

# Visualizar os dados das características 0 e 1.
plt.figure(figsize=(8, 6))

for i in range(len(wine.target_names)):
    plt.scatter(X[y == i][:, 0], X[y == i][:, 1], label=wine.target_names[i])

plt.xlabel('Característica 1 - Álcool')
plt.ylabel('Característica 2 - Ácido Málico')
plt.title('Visualização dos Dados Wine')
plt.legend()
plt.show()
```

### **5.**   Decision Tree:

```python
# Importações de bibliotecas:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Carregamento da base de dados WINE:
banco_de_dados = load_wine()
wine_df = pd.DataFrame(banco_de_dados.data, columns=banco_de_dados.feature_names)

# Adição da coluna 'classe' ao DataFrame:
wine_df['classe'] = banco_de_dados['target']

# Armazenamento dos rótulos de classe em uma variável:
wine_classe = wine_df['classe']

# Remoção da coluna 'classe' do DataFrame, porque ela representa a variável de resposta ou saída do sistema e não deve ser usada como atributo de entrada para o modelo.
wine_df.drop(['classe'], axis=1, inplace=True)

# Divisão dos dados em conjuntos de treinamento e teste:
X_train, X_test, y_train, y_test = train_test_split(wine_df, wine_classe, test_size=0.4, random_state=13)

# Criação do modelo de árvore de decisão:
arvore_de_decisao = DecisionTreeClassifier(random_state=13)

# Treinamento do modelo:
arvore_de_decisao.fit(X_train, y_train)

# Previsões com o modelo treinado:
y_previsao = arvore_de_decisao.predict(X_test)

# Avaliação do desempenho do modelo:
accuracy = accuracy_score(y_test, y_previsao)
conf_matrix = confusion_matrix(y_test, y_previsao)
class_report = classification_report(y_test, y_previsao, target_names=banco_de_dados.target_names)

# Impressão dos resultados:
print(f'Acurácia: {accuracy:.2f}\n')
print('Matriz de Confusão:\n')
print(conf_matrix)
print('\n')
print('Relatório de Classificação:\n')
print(class_report)
print('\n')
fig, ax = plt.subplots(figsize=(10,6))
tree.plot_tree(arvore_de_decisao)
print('Visualização Esquemática:\n')
plt.show()
```

### **6.**   Random Forest:

Esta técnica cria árvores de decisão em amostras de dados e então obtém a previsão de cada uma delas, selecionando finalmente a melhor solução por meio de votação.

É um método de conjunto (ensemble method) que é superior a uma única árvore de decisão, pois reduz o overfitting pela média dos resultados.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Carregar o conjunto de dados de vinhos.
wine = load_wine()
X = wine.data
y = wine.target

# Dividir os dados em conjuntos de treino e teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=13)

# Criar o classificador Random Forest.
clf = RandomForestClassifier(n_estimators=5, random_state=13)

# Treinar o modelo.
clf.fit(X_train, y_train)

# Fazer previsões no conjunto de teste.
y_pred = clf.predict(X_test)

# Calcular e exibir a acurácia.
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia do Random Forest: {accuracy:.2f}')
print('\n')

# Exibir relatório de classificação.
relatorio_de_classificacao = classification_report(y_test, y_pred, target_names=wine.target_names)
print(relatorio_de_classificacao)
```

### **7.**   Linear Regression:

A técnica em foco é a regressão linear, usada para estabelecer a relação entre uma variável dependente e variáveis independentes, buscando identificar uma relação linear entre elas.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Carregar o conjunto de dados de vinhos.
wine = load_wine()
X = wine.data
y = wine.target

# Dividir os dados em conjuntos de treino e teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=13)

# Criar o modelo de regressão linear.
model = LinearRegression()

# Treinar o modelo
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste.
y_pred = model.predict(X_test)

# Arredondar as previsões para obter classes (0, 1, 2) -> Uma vez que os valores armazenados na variável "y_pred" são contínuos.
y_pred_aproximado = [round(pred) for pred in y_pred]

# Calcular e exibir a acurácia (não apropriado para classificação) -> Se não usar os valores aproximados, não é possível determinar a acurácia do método.
accuracy = accuracy_score(y_test, y_pred_aproximado)
print(f'Acurácia da Regressão Linear para Classificação: {accuracy:.2f}')
print('\n')

# Observar a matriz confusão produzida pela máquina.
matriz_confusao = confusion_matrix(y_test, y_pred_aproximado)
print('Matriz de Confusão:\n')
print(matriz_confusao)
```

### **8.**   K-means Algorithm:

O K-means é um algoritmo de clustering que, ao iterar para encontrar centróides ótimos, busca minimizar a distância entre os pontos de dados e esses centróides.

Essa abordagem é útil para identificar padrões e estruturas nos dados, sendo especialmente eficaz quando o número de clusters é conhecido a priori.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Carregar o conjunto de dados de vinhos.
wine = load_wine()
X = wine.data

# Padronizar os dados para ter média zero e variância unitária.
normalizador = StandardScaler()
X_normalizado = normalizador.fit_transform(X)

# Criar o modelo K-means com 3 clusters (número de classes na base de dados Wine) -> "n_init" controla o número de vezes que o algoritmo KMeans será executado com diferentes centroides iniciais.
kmeans = KMeans(n_clusters=3, random_state=13, n_init=10)

# Treinar o modelo K-means.
kmeans.fit(X_normalizado)

# Visualização dos clusters (usaremos duas características para visualização).
for i in range(len(wine.target_names)):
    plt.scatter(X[y == i][:, 4], X[y == i][:, 9], label=wine.target_names[i], edgecolors='black')

plt.xlabel('Característica 4 - Magnésio')
plt.ylabel('Característica 9 - Intensidade da cor')
plt.title('K-means Clustering nos Dados da Base Wine')
plt.legend()
plt.show()
```

### **9.**   Hierarchical Clustering:

O Hierarchical Clustering, ou Clustering Hierárquico, é um método de agrupamento de dados que organiza os elementos em uma estrutura de árvore (dendrograma).

Nos algoritmos hierárquicos aglomerativos, cada ponto de dado é considerado um cluster individual e, em seguida, são mesclados ou aglomerados progressivamente (abordagem de baixo para cima) os pares de clusters. A hierarquia dos clusters é mostrada como um dendrograma ou estrutura de árvore.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
import scipy.cluster.hierarchy as shc

# Carregar o conjunto de dados de vinhos.
wine = load_wine()
X = wine.data

# Padronizar os dados para ter média zero e variância unitária.
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Visualizar o dendrograma.
plt.figure(figsize=(10, 6))
dend = shc.dendrogram(shc.linkage(X_normalized, method='ward'))
plt.title('Dendrograma Hierárquico dos Dados Wine')
plt.xlabel('Amostras')
plt.ylabel('Distância')
plt.show()
```

### **11.**   Improving Performance of ML Models:

Os **"Ensembles"** podem proporcionar um impulso no resultado da aprendizagem automática através da **combinação de vários modelos**.

Basicamente, os modelos de "ensembles" consistem em vários modelos de aprendizagem supervisionada treinados individualmente, tendo os seus resultados combinados de diversas formas para obter um melhor desempenho de previsão em comparação com um único modelo de aprendizagem de máquina.

Abaixo está o algoritmo **AdaBoost**, um dos algoritmos de reforço mais bem sucedidos.

A chave principal deste algoritmo está na forma como atribui pesos às instâncias do conjunto de dados.

Devido a este fato, o algoritmo precisa prestar menos atenção às instâncias enquanto constrói os modelos subsequentes.

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier

# Carregamento da base de dados Wine.
dadosWine = load_wine()
dframe = pd.DataFrame(data = dadosWine.data, columns = dadosWine['feature_names'])
dframe['target'] = dadosWine.target

# Visualizar o dataframe gerado.
dframe
```

```python
# Desenvolvimento do método com base nas informações do material de apoio.
x = dframe.drop('target', axis=1)
y = dframe.target

seed = 5
kfold = KFold(n_splits=10, random_state=None)
num_trees = 50
model = AdaBoostClassifier(n_estimators=num_trees, random_state=seed)
results = cross_val_score(model, x, y, cv=kfold)

# Exibir os resultados.
print(f'O resultado obtido no aprendizado de máquina foi de: {results.mean():.2}')
```
